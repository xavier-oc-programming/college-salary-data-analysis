## **4)— Accessing Columns and Individual Cells in a DataFrame**

### **Objective**

Learn how to:

* Access individual columns
* Compute summary statistics
* Identify rows associated with extreme values
* Retrieve individual cells and entire rows using index-based selection

---

### **Accessing a Single Column**

To access a specific column in a DataFrame, use **square bracket notation** with the column name:

```python
clean_df['Starting Median Salary']
```

This returns a **Pandas Series** containing all values from that column.

Use this approach whenever you want to:

* Inspect values
* Perform calculations
* Chain additional methods

---

### **Finding the Maximum Value in a Column**

To find the highest starting salary, chain the `.max()` method:

```python
clean_df['Starting Median Salary'].max()
```

This returns the **maximum numeric value** in that column.

---

### **Finding the Index of the Maximum Value**

To determine *which row* contains that maximum value, use `.idxmax()`:

```python
clean_df['Starting Median Salary'].idxmax()
```

This returns the **index label** of the row with the largest value.

In this dataset, the result is:

```text
43
```

---

### **Accessing a Single Cell**

Once you know the index, you can retrieve the corresponding value from another column using `.loc`:

```python
clean_df['Undergraduate Major'].loc[43]
```

This returns the **college major** associated with the highest starting salary:

```text
'Physician Assistant'
```

---

### **Alternative: Double Bracket Indexing**

You may also see this written as:

```python
clean_df['Undergraduate Major'][43]
```

While this works, using `.loc` is **preferred** because it is:

* More explicit
* Safer for complex indexing
* Consistent with Pandas best practices

---

### **Retrieving an Entire Row**

If you omit the column selection and use `.loc` with only the index, Pandas returns the **entire row**:

```python
clean_df.loc[43]
```

This outputs:

* Undergraduate Major
* Starting Median Salary
* Mid-Career Median Salary
* Salary percentiles
* Degree group (STEM, Business, HASS)

The result is a **Series** representing all information for that major.

---

### **Key Concepts Reinforced**

* Columns are accessed with `df['column_name']`
* `.max()` returns the highest value
* `.idxmax()` returns the index of that value
* `.loc[row_index, column_name]` retrieves a specific cell
* `.loc[row_index]` retrieves an entire row
