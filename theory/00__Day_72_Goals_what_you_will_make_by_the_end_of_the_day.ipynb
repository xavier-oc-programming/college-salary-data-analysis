## **0)- Day 72 Goals - What You Will Make by the End of the Day**

### **Introduction**

In this lesson, you will learn **advanced data exploration with Pandas** by analyzing **post-university salaries of graduates by college major**.

College degrees are expensive, but do they pay off equally?  
Using real-world salary data collected by **PayScale Inc.** from a survey of **1.2 million Americans with bachelor’s degrees**, we will explore how earnings differ across majors and degree categories.

This lesson focuses on **asking meaningful questions of data** and using Pandas to answer them systematically.

---

### **Project Context**

The dataset allows us to investigate salary outcomes across different college majors and career paths.

We will use Pandas to:
- Load and inspect real salary data
- Clean missing or invalid values
- Extract specific information
- Compare earnings across majors and degree groups

---

### **Key Questions We Will Answer**

- Which degrees have the **highest starting salaries**?
- Which majors have the **lowest earnings after college**?
- Which degrees have the **highest earning potential**?
- Which college majors represent the **lowest risk** from an earnings standpoint?
- Do **Business**, **STEM**, or **HASS** (Humanities, Arts, Social Sciences) degrees earn more on average?

---

### **What You Will Learn Today**

- How to explore a **Pandas DataFrame**
- How to detect **NaN (Not a Number)** values and clean your data
- How to select specific **columns**, **rows**, and **individual cells**
- How to sort your data to reveal rankings and patterns
- How to group data by category
- How to pivot datasets to produce structured summaries

---

### **Core Pandas Skills Reinforced**

- `read_csv()`
- `.head()`, `.info()`, `.describe()`
- `.isna()`, `.dropna()`
- Column and row selection
- `.sort_values()`
- `.groupby()`
- `.pivot_table()`

---

### **Learning Outcome**

By the end of Day 72, you will be able to take a **real-world dataset**, formulate **clear analytical questions**, and use **Pandas** to extract meaningful, career-relevant insights.
