## **6)—Sorting Values & Adding Columns: Majors with the Most Potential vs Lowest Risk**

### **Concept: What “Risk” Means in Salary Data**

A **low-risk major** is one where earnings are more predictable.
In data terms, this means a **small difference** between:

* Mid-Career **10th Percentile Salary** (lower earners)
* Mid-Career **90th Percentile Salary** (top earners)

A smaller gap implies:

* Less income variability
* More predictable career outcomes

---

### **Calculating Salary Spread**

To measure this risk, we calculate the **spread**:

```
Spread = 90th percentile salary − 10th percentile salary
```

Because Pandas supports vectorized operations, we can subtract entire columns directly.

```python
spread_col = (
    clean_df['Mid-Career 90th Percentile Salary']
    - clean_df['Mid-Career 10th Percentile Salary']
)
```

---

### **Alternative: Using `.subtract()`**

This does the same thing, but is sometimes clearer when chaining methods:

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(
    clean_df['Mid-Career 10th Percentile Salary']
)
```

Both approaches are valid and produce identical results.

---

### **Adding the Spread as a New Column**

We now insert the computed spread into the dataframe.

```python
clean_df.insert(1, 'Spread', spread_col)
```

#### Why `insert(1, ...)`?

* The first argument (`1`) specifies **column position**
* Position `1` places `Spread` as the **second column**
* This is optional, but improves readability

---

### **Inspecting the Updated DataFrame**

```python
clean_df.head()
```

You should now see a new column called **Spread**, representing salary variability for each major.

---

### **Sorting by Lowest Risk (Smallest Spread)**

To identify majors with the **smallest salary spread**, we sort in ascending order:

```python
low_risk = clean_df.sort_values('Spread')
```

By default, `.sort_values()` sorts **ascending**, which is exactly what we want here.

---

### **Viewing the Lowest-Risk Majors**

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

This shows:

* Majors with the **most predictable earnings**
* The smallest difference between low and high earners

---

### **Key Takeaways**

* Column arithmetic in Pandas is **vectorized** and efficient
* `.insert()` allows precise control over column placement
* `.sort_values()` defaults to ascending order
* Smaller salary spread = **lower earnings risk**
* This technique combines:

  * Feature engineering
  * Sorting
  * Subsetting columns
