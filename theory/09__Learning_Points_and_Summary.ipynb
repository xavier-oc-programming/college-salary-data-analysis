## **09)—Learning Points & Summary**

### **Overview**

This final section consolidates the key Pandas concepts and techniques used throughout **Day 72**.
The focus has been on **exploratory data analysis**, **cleaning**, **comparison**, and **aggregation** using real-world salary data.

---

### **Exploring a DataFrame**

You learned how to inspect the structure and contents of a DataFrame using:

* `.head()` and `.tail()`

  * Preview the first or last rows of the dataset.
* `.shape`

  * Identify the number of rows and columns.
* `.columns`

  * Retrieve all column names.

These methods help you understand what data you are working with before analysis.

---

### **Handling Missing Data**

To detect and clean missing values:

* `.isna()`

  * Identify NaN (Not a Number) values in the DataFrame.
* `.dropna()`

  * Remove rows containing missing values when appropriate.

This ensures that calculations and aggregations are reliable.

---

### **Accessing Data**

You practiced multiple ways to retrieve data:

* **Columns**

  * `df['column_name']`
  * `df[['column_1', 'column_2']]`
* **Individual cells**

  * `df['column_name'][index]`
  * `df.loc[index, 'column_name']`

Using `.loc` is preferred for clarity and robustness.

---

### **Finding Extreme Values**

To locate highest and lowest values and their positions:

* `.max()` and `.min()`

  * Find extreme values.
* `.idxmax()` and `.idxmin()`

  * Find the index (row label) where those extremes occur.

These methods were used to identify majors with the highest and lowest salaries.

---

### **Sorting and Feature Engineering**

You learned how to:

* Sort data using `.sort_values()`
* Create new columns using `.insert()` or column arithmetic

This enabled comparisons such as:

* Lowest-risk vs highest-risk majors
* High earning potential vs salary stability

---

### **Grouping and Aggregation**

To perform Excel-style pivot analysis:

* `.groupby()`

  * Group rows by a categorical column (e.g. `Group`)
* `.mean(numeric_only=True)`

  * Compute average values safely across numeric columns

This allowed comparison of salary outcomes across degree categories such as **STEM**, **Business**, and **HASS**.

---

### **Key Takeaways**

* Pandas enables efficient, readable analysis of real-world datasets.
* Clean data is essential before drawing conclusions.
* Group-level insights often reveal trends hidden at the individual level.
* Combining indexing, aggregation, and sorting is a powerful analytical pattern.

---

This concludes **Day 72 – Advanced Data Exploration with Pandas**.
