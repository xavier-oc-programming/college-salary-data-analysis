## **3)— Preliminary Data Exploration and Data Cleaning with Pandas**

### **Introduction**

Once the data has been loaded into a Pandas DataFrame, the first step in any data science project is to **understand the structure and quality of the data**.
This stage focuses on answering basic but critical questions about the dataset before performing any analysis.

---

### **Questions to Answer First**

* How many **rows** does the DataFrame have?
* How many **columns** does it have?
* Do the columns have clear **names and labels**?
* Are there any **missing values** or **junk rows** that should be removed?

---

### **Inspecting the Size of the DataFrame**

To determine the number of rows and columns, use the `shape` attribute:

```python
df.shape
```

* Returns a tuple: `(number_of_rows, number_of_columns)`
* In this dataset, the output is:

```text
(51, 6)
```

This tells us:

* 51 rows
* 6 columns

---

### **Inspecting Column Names**

To view all column labels, use the `columns` attribute:

```python
df.columns
```

This confirms:

* Each column is properly named
* Column labels describe salary metrics and degree grouping

Understanding column names is essential before selecting or manipulating data.

---

### **Checking for Missing or Junk Data**

Before analysis, it is important to identify **NaN (Not a Number)** values.
NaN values represent:

* Blank cells
* Invalid numeric data
* Placeholder rows (e.g. metadata or source notes)

To locate missing values, use:

```python
df.isna()
```

This returns a boolean DataFrame indicating where values are missing.

---

### **Inspecting the Bottom of the DataFrame**

To quickly spot suspicious rows, inspect the last few rows:

```python
df.tail()
```

This reveals a final row that:

* Contains source information (`PayScale Inc.`)
* Has `NaN` values for all salary-related columns

This row is **not part of the actual dataset** and should be removed.

---

### **Removing the Junk Row**

There are two possible approaches:

* Manually drop the row by index
* Automatically remove rows containing missing values

The simplest and safest approach here is to use `dropna()`:

```python
clean_df = df.dropna()
```

To confirm the row has been removed:

```python
clean_df.tail()
```

The final row should now contain valid degree data only.

---

### **Resulting Clean Dataset**

After cleaning:

* The DataFrame no longer contains junk or metadata rows
* All remaining rows represent valid salary information
* The dataset is now safe to use for sorting, grouping, and analysis

---

### **Key Takeaways**

* Always inspect **shape** and **columns** before analysis
* Use `isna()` and `tail()` to identify hidden data issues
* Remove invalid rows early to avoid misleading results later
* Data cleaning is a **mandatory first step**, not an optional one
