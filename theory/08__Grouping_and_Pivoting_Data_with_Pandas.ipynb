## **8)—Grouping and Pivoting Data with Pandas**

### **Objective**

Move from **individual majors** to **degree categories** (`STEM`, `HASS`, `Business`) and compare them using aggregated statistics — similar to an Excel Pivot Table.

---

### **Using `.groupby()`**

The `Group` column categorizes each major.
By grouping on this column, we can compute summary statistics per category.

```python
clean_df.groupby('Group')
```

On its own, this creates a grouped object. To get results, we apply an **aggregation method**.

---

### **Counting Majors per Group**

```python
clean_df.groupby('Group').count()
```

#### Interpretation

* One row per group
* Counts represent how many **non-null values** exist per column
* Since the dataset is clean, this reflects the number of majors in each category

---

### **Mini Challenge — Average Salary by Group**

**Question:**
Which degree category has the highest *average* salary?

To answer this, we compute the mean for each numeric column **per group**.

```python
clean_df.groupby('Group').mean(numeric_only=True)
```

#### Why `numeric_only=True`?

* Some columns (e.g. major names) are non-numeric
* This argument ensures Pandas only averages salary-related columns
* Prevents `TypeError` when object columns are present

---

### **Formatting the Output (2 Decimal Places)**

To make the table readable in the notebook:

```python
pd.options.display.float_format = '{:,.2f}'.format
```

Now rerun:

```python
clean_df.groupby('Group').mean(numeric_only=True)
```

#### Result

* Salaries are displayed with commas and two decimals
* Underlying data remains unchanged

---

### **How to Read the Resulting Table**

Each row represents a **degree category** (`Business`, `HASS`, `STEM`).

Each column shows the **average** of that salary metric for the group, for example:

* Average starting salary
* Average mid-career median salary
* Average 10th / 90th percentile salaries
* Average spread (risk)

This allows direct comparison of earning potential **by category**, not by individual major.

---

### **Key Takeaways**

* `.groupby()` + `.mean()` is the Pandas equivalent of an Excel Pivot Table
* `numeric_only=True` avoids aggregation errors
* Formatting affects **display only**, not the data
* Group-level analysis reveals trends hidden at the individual-major level
