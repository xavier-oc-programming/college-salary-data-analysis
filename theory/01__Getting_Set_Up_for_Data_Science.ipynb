## **1)— Getting Set Up for Data Science**

### **Introduction**

When exploring and visualising datasets, the **Python Notebook format** is better suited than a traditional IDE such as PyCharm. Notebooks allow you to execute code in small steps and immediately see outputs, which is ideal for data analysis and experimentation.

In this lesson, you are introduced to the **Google Colab Notebook**, an online environment for running Jupyter-style notebooks.

---

### **Using Google Colab**

Google Colab allows you to create and run Python notebooks directly in your browser, with no local setup required.

To create a new notebook:

* Open **Google Drive**
* Click **New → More → Google Colaboratory**
* A new Python notebook (`.ipynb`) will be created and saved to your Drive

Google Colab functions as an **online version of Jupyter Notebook**.

---

### **Local Alternative: Anaconda + Jupyter**

If Google Colab is unavailable or you prefer to work locally:

* Install **Anaconda**
* Use the bundled **Jupyter Notebook** included with Anaconda

Both Google Colab and local Jupyter notebooks work the same way from a user perspective.

---

### **How a Python Notebook Works**

A notebook is divided into **cells**:

* Each cell can be executed independently
* Output appears directly below the cell

To run a cell:

* Use the keyboard shortcut **Shift + Enter**

This structure supports:

* Incremental exploration
* Visual inspection of data
* Mixing code, text, and results in one place

---

### **Runtime Connection**

Before executing any code, the notebook must connect to a **runtime**:

* Google Colab automatically prompts you to connect
* The runtime provides the environment where your code executes

Without a connected runtime, code cells will not run.

---

### **Key Takeaway**

Python notebooks are the standard environment for **data science and data analysis**.
They enable fast iteration, immediate feedback, and clear documentation of your analytical process.
